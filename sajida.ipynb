```python
x = dataset.drop('Has_Asthma', axis=1)
y = dataset['Has_Asthma']
```

## Split data

### Subtask:
Split the data into training and testing sets.


**Reasoning**:
Split the data into training and testing sets as requested in the instructions.



```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(
    x, y, random_state=42, test_size=0.2
)
```

## Scale numerical features

### Subtask:
Use StandardScaler to scale the numerical columns in the training and testing sets.


**Reasoning**:
Identify the numerical columns in the training data and scale them using StandardScaler, then apply the same scaling to the test data.



```python
from sklearn.preprocessing import StandardScaler

numerical_cols = xtrain.select_dtypes(include=['int64', 'float64']).columns
scaler = StandardScaler()

xtrain[numerical_cols] = scaler.fit_transform(xtrain[numerical_cols])
xtest[numerical_cols] = scaler.transform(xtest[numerical_cols])

display(xtrain.head())
display(xtest.head())
```

### Subtask:
Choose a few classification models to evaluate. Some common ones for this type of problem include Logistic Regression, Decision Tree Classifier, Random Forest Classifier, Support Vector Machine (SVM), and K-Nearest Neighbors (KNN).


## Train and evaluate models

### Subtask:
Train each selected model on the *scaled* training data and evaluate their performance on the *scaled* testing data using appropriate metrics (e.g., accuracy, precision, recall, F1-score).


**Reasoning**:
Train and evaluate the selected classification models on the scaled data.



```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Initialize and train models
log_reg_model = LogisticRegression(random_state=42)
dt_model = DecisionTreeClassifier(random_state=42)
rf_model = RandomForestClassifier(random_state=42)
svm_model = SVC(random_state=42)
knn_model = KNeighborsClassifier()

log_reg_model.fit(xtrain, ytrain)
dt_model.fit(xtrain, ytrain)
rf_model.fit(xtrain, ytrain)
svm_model.fit(xtrain, ytrain)
knn_model.fit(xtrain, ytrain)

# Make predictions
y_pred_log_reg = log_reg_model.predict(xtest)
y_pred_dt = dt_model.predict(xtest)
y_pred_rf = rf_model.predict(xtest)
y_pred_svm = svm_model.predict(xtest)
y_pred_knn = knn_model.predict(xtest)

# Evaluate models
accuracy_log_reg = accuracy_score(ytest, y_pred_log_reg)
precision_log_reg = precision_score(ytest, y_pred_log_reg)
recall_log_reg = recall_score(ytest, y_pred_log_reg)
f1_log_reg = f1_score(ytest, y_pred_log_reg)

accuracy_dt = accuracy_score(ytest, y_pred_dt)
precision_dt = precision_score(ytest, y_pred_dt)
recall_dt = recall_score(ytest, y_pred_dt)
f1_dt = f1_score(ytest, y_pred_dt)

accuracy_rf = accuracy_score(ytest, y_pred_rf)
precision_rf = precision_score(ytest, y_pred_rf)
recall_rf = recall_score(ytest, y_pred_rf)
f1_rf = f1_score(ytest, y_pred_rf)

accuracy_svm = accuracy_score(ytest, y_pred_svm)
precision_svm = precision_score(ytest, y_pred_svm)
recall_svm = recall_score(ytest, y_pred_svm)
f1_svm = f1_score(ytest, y_pred_svm)

accuracy_knn = accuracy_score(ytest, y_pred_knn)
precision_knn = precision_score(ytest, y_pred_knn)
recall_knn = recall_score(ytest, y_pred_knn)
f1_knn = f1_score(ytest, y_pred_knn)

# Print evaluation metrics
print("Logistic Regression Performance:")
print(f"Accuracy: {accuracy_log_reg:.4f}")
print(f"Precision: {precision_log_reg:.4f}")
print(f"Recall: {recall_log_reg:.4f}")
print(f"F1-Score: {f1_log_reg:.4f}")
print("\nDecision Tree Performance:")
print(f"Accuracy: {accuracy_dt:.4f}")
print(f"Precision: {precision_dt:.4f}")
print(f"Recall: {recall_dt:.4f}")
print(f"F1-Score: {f1_dt:.4f}")
print("\nRandom Forest Performance:")
print(f"Accuracy: {accuracy_rf:.4f}")
print(f"Precision: {precision_rf:.4f}")
print(f"Recall: {recall_rf:.4f}")
print(f"F1-Score: {f1_rf:.4f}")
print("\nSVM Performance:")
print(f"Accuracy: {accuracy_svm:.4f}")
print(f"Precision: {precision_svm:.4f}")
print(f"Recall: {recall_svm:.4f}")
print(f"F1-Score: {f1_svm:.4f}")
print("\nKNN Performance:")
print(f"Accuracy: {accuracy_knn:.4f}")
print(f"Precision: {precision_knn:.4f}")
print(f"Recall: {recall_knn:.4f}")
print(f"F1-Score: {f1_knn:.4f}")
```
